```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3],[2,9,15])
plt.title("Bikash's Data")
plt.xlabel("This is X")
plt.ylabel("This is Y")
plt.show()
```

```python
# You can add more styles see documentation matplotlib styles
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[2,3,4]
y=[7,6,5]
x1=[5,6,4]
y1=[6,5,3]

plt.plot(x,y,label="first",linewidth=3)
plt.plot(x1,y1,label="Second")
plt.legend()                             # Very important as it shows the upper-right corner text 
plt.show 
```
